```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/IMDB_Top250Engmovies2_OMDB_Detailed.csv')
df.head()
```

```python
df = df[['Title','Genre','Director','Actors','Plot']]
df.head()
```

```python
from rake_nltk import Rake
import nltk
nltk.download('stopwords')
nltk.download('punkt')
# initializing the new column
df['Key_words'] = " "
for index, row in df.iterrows():
  plot = row['Plot']
  r = Rake()
  r.extract_keywords_from_text(plot)
  key_words_dict_scores = r.get_word_degrees()
  row['Key_words'] = list(key_words_dict_scores.keys())
```

```python
df.head()
```

```python
for index,row in df.iterrows():
  row['Genre'] = row['Genre'].split(', ')
  row['Director'] = "".join(row['Director'].split(' '))
  row['Actors'] = [ "".join(x.split(" ")) for x in row['Actors'].split(', ')[:3] ]
df.head()
df['Director'] = df['Director'].apply(lambda x:x.split()) #converting string into list
df['tags'] = df['Genre'] + df['Director'] + df['Actors'] + df['Key_words']
```

```python
df.head()
```

```python
new_df = df.drop(columns=['Genre', 'Director','Key_words','Actors'])
```

```python
#new_df['tags'] = new_df['tags'].apply(lambda x: "".join(x))
new_df.head() 
```

```python
new_df.drop('Plot',axis=1,inplace=True)
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer()
count_matrix = count.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(count_matrix)
```

```python
def recommendations(movie):
  recommended_movies = []
  idx = new_df[new_df['Title'] == movie].index[0] 
  similar_mov = cosine_sim[idx]
  top_10_indexes = sorted(list(enumerate(similar_mov)), reverse=True, key= lambda x: x[1])[1:11]
  for i in top_10_indexes:
    print(new_df.iloc[i[0]].Title)
```

```python
print(recommendations('Shutter Island'))
```
